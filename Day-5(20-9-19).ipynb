### turtle

```python
# design sprial square
import turtle as t
a1=t.Turtle()
a1.pencolor('red')
for i in range(200):
    a1.forward(i)
    a1.left(91)
a1.done()
```

```python
# design spiral star
import turtle as t
a1=t.Turtle()
a1.pencolor('blue')
for i in range(50):
    a1.forward(i*10)
    a1.left(144)
t.done()    
```

```python
# design hexagon
from turtle import *
pencolor('blue')
colors=['blue','green','red','yellow','orange','purple']
for i in range(360):
    pencolor(colors[i%6])
    width(i/100+1)
    forward(i)
    left(59)
```

```python
# design olympics logo
from turtle import *
pensize(6)
r1 = ['blue','black','red']
for i in range(3):
    penup()
    pencolor(r1[i%3])
    goto(i*100,0)
    pendown()
    circle(45)
r2=['','yellow','','green']
for i in range(1,4,2):
    penup()
    pencolor(r2[i])
    goto(i*55,-55)
    pendown()
    circle(50)
```

###  Lambda functions
 - anonymous functions means that a function is without a name
 - any lamba function on python will be defined with lambda
 - syntax:- lambda arg: expression

```python
def square(n):
    return n*n
square(10)
```

```python
a= lambda x :x*x
print(a(10))
```

### use of lambda with filter

```python
def filterList(li):
    a=[]
    for i in li:
        if i%2==0:
            a.append(i)
    return a
li=[1,2,3,4,5,6,7,8]
filterList(li)
```

```python
li =[1,2,3,4,5,6,7,8]
lam_li=list(filter(lambda x:(x%2==0),li))
print(lam_li)
```

### lambda with map()
 - map() function in python takes in a function and list as arguments

```python
li =[1,2,3,4,5,6,7,8]
lam_li=list(map(lambda x:(x*2),li))
print(lam_li)
```
